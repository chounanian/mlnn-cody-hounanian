### 1. Show the most common words in the article.


```python
from collections import Counter

words = [token.text
         for token in processed_text
         if not token.is_stop and not token.is_punct]

word_freq = Counter(words)
common_words = word_freq.most_common(20)

import pandas as pd

print("\nTOP 20 USED WORDS:")
print(pd.DataFrame(common_words))
```

### 2. Show the most common words under a part of speech. (i.e. NOUN: {'Bob':12, 'Alice':4,})

```python
# NOUN tokens
nouns = [token.text
         for token in processed_text
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "NOUN")]

# Most common NOUN tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(10)


# VERB tokens
verbs = [token.text
         for token in processed_text
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "VERB")]

# Most common VERB tokens
verb_freq = Counter(verbs)
common_verbs = verb_freq.most_common(10)


# PROPN tokens
propnouns = [token.text
         for token in processed_text
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "PROPN")]

# Most common VERB tokens
propnoun_freq = Counter(propnouns)
common_propnouns = propnoun_freq.most_common(10)

import pandas as pd

print("\nCOMMON NOUNS:")
print(pd.DataFrame(common_nouns))

print("\nCOMMON VERBS:")
print(pd.DataFrame(common_verbs))

print("\nCOMMON PROPER NOUNS:")
print(pd.DataFrame(common_propnouns))
```

### 3. Find a subject/object relationship through the dependency parser in any sentence.

```python
# Using Parser

def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)

for i in range(1,1000,1):
    for sent in processed_text.sents:
        i += 1
        if i == 1000:
            tree = sent
            tree_sent = pr_tree(tree.root,0)
            print(tree_sent)
```

```python
# Printing the Subject/Object Pairs

def get_subject_phrase(doc):
    for token in doc:
        if ("subj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]

        
def get_object_phrase(doc):
    for token in doc:
        if ("dobj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]

sents = list(processed_text.sents)

for sentence in sents:
    doc = processor(str(sentence))
    subject_phrase = get_subject_phrase(doc)
    object_phrase = get_object_phrase(doc)
    print("\nSUBJECT:")
    print(subject_phrase)
    print("OBJECT:")
    print(object_phrase)
```

### 4. Show the most common Entities and their types. 

```python
ents = Counter()
for ent in processed_text.ents:
    ents[f"{ent.label_}:{ent.text}"] += 1

common_ent = ents.most_common(10)
print("\nCOMMON ENTITIES:")
print(pd.DataFrame(common_ent))
```

### 5. Find Entites and their dependency (hint: entity.root.head)

```python
# Using the Parser

for entity in processed_text.ents:
    tree = pr_tree(entity.root.head, 0)
print(tree)
```

```python
# Printing Dependency List

for ent in processed_text.ents:
    print("ENTITY:", ent.text, "\nDEPENDENCY:", ent.root.head, "\n")
```
