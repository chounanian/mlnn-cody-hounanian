## Part 1 - Answer


### 2 Neurons Per Hidden Layer

```python
from keras.utils.tf_utils import set_random_seed
set_random_seed(7) # for consistency

num_layers = [1,2,3,4,5]
scores = []
layers = []
sgd = SGD(lr=0.1)

for num_layer in num_layers:
    
    model = Sequential()
    
    # build model and evaluate 
    model.add(Dense(2, input_dim=2, activation='tanh')) #input layer
    
    for i in range(num_layer - 1):
        model.add(Dense(2, input_dim=2, activation='tanh')) #hidden layers

    model.add(Dense(1, input_dim=2, activation='tanh')) #output layer

    model.compile(loss='binary_crossentropy', optimizer='sgd')
    history = model.fit(X, y, batch_size=10, epochs=400) #160/4 = 40 per epoch
    
    # evaluate the model
    loss_scores = history.history["loss"]
    scores.append(loss_scores)
    layer = num_layer
    layers.append(layer)
    epoch_num = list(range(1, len(loss_scores)+1))


# plot scores
plot1 = plt.figure()

for i in range(len(scores)):
    plt.plot(epoch_num, scores[i], label = i+1)
plt.legend(title = "Number of Layers", loc = 'upper right')
plt.xlabel("Epochs")
plt.ylabel("Loss Scores")
plt.title("2 Neurons Per Hidden Layer")
```

### 4 Neurons Per Layer and 5 Hidden Layers

```python
set_random_seed(7) # for consistency

activations_list = ["tanh", "softplus", "relu", "sigmoid", "softmax", "selu", "elu"]
scores = []
layers = []
sgd = SGD(lr=0.1)

for activations in activations_list:
    
    model = Sequential()
    
    # build model and evaluate 
    model.add(Dense(2, input_dim=2, activation='tanh')) #input layer
    model.add(Dense(4, input_dim=2, activation=activations)) #hidden layer
    model.add(Dense(4, input_dim=2, activation=activations)) #hidden layer
    model.add(Dense(4, input_dim=2, activation=activations)) #hidden layer
    model.add(Dense(4, input_dim=2, activation=activations)) #hidden layer
    model.add(Dense(4, input_dim=2, activation=activations)) #hidden layer
    model.add(Dense(1, input_dim=2, activation='tanh')) #output layer
    model.compile(loss='binary_crossentropy', optimizer='sgd')
    history = model.fit(X, y, batch_size=10, epochs=400) #160/4 = 40 per epoch
    
    # evaluate the model
    loss_scores = history.history["loss"]
    scores.append(loss_scores)
    layer = 3
    layers.append(layer)
    epoch_num = list(range(1, len(loss_scores)+1))


# plot scores
plot2 = plt.figure()

for i in range(len(scores)):
    plt.plot(epoch_num, scores[i], label = activations_list[i])
plt.legend(title = "Activations", loc = 'upper right')
plt.xlabel("Epochs")
plt.ylabel("Loss Scores")
plt.title("5 Neurons Per Layer and 5 Hidden Layers")
```

### 4 Neurons Per Layer, 5 Hidden Layers, and TANH Activation

```python
from keras.optimizers import Adam
from keras.optimizers import Nadam
from keras.optimizers import RMSprop
from keras.optimizers import Adadelta

set_random_seed(7) # for consistency

sgd = SGD(learning_rate=0.001)
adam = Adam(learning_rate=0.001)
nadam = Nadam(learning_rate=0.001)
rmsprop = RMSprop(learning_rate=0.001)
adadelta = Adadelta(learning_rate=0.001)

optimizers_list = [sgd, adam, nadam, rmsprop, adadelta]
optimizers_names = ["SGD", "Adam", "Nadam", "RMSprop", "Adadelta"]

scores = []
layers = []

for optimizers in optimizers_list:
    
    model = Sequential()
    
    # build model and evaluate 
    model.add(Dense(2, input_dim=2, activation='tanh')) #input layer
    model.add(Dense(4, input_dim=2, activation= 'tanh')) #hidden layer
    model.add(Dense(4, input_dim=2, activation= 'tanh')) #hidden layer
    model.add(Dense(4, input_dim=2, activation= 'tanh')) #hidden layer
    model.add(Dense(4, input_dim=2, activation= 'tanh')) #hidden layer
    model.add(Dense(1, input_dim=2, activation='tanh')) #output layer
    model.compile(loss='binary_crossentropy', optimizer=optimizers)
    history = model.fit(X, y, batch_size=10, epochs=400) #160/4 = 40 per epoch
    
    # evaluate the model
    loss_scores = history.history["loss"]
    scores.append(loss_scores)
    layer = 3
    layers.append(layer)
    epoch_num = list(range(1, len(loss_scores)+1))


# plot scores
plot2 = plt.figure()

for i in range(len(scores)):
    plt.plot(epoch_num, scores[i], label = optimizers_names[i])
plt.legend(title = "Optimizers", loc = 'upper right')
plt.xlabel("Epochs")
plt.ylabel("Loss Scores")
plt.title("4 Neurons Per Layer, 5 Hidden Layers, and tanh Activation")
```

After reviewing the previous models, it appears that a 4 neuron per hidden layer neural network that features 5 hidden layers and utilizes a tanh activation is the most effective. This is supported by the fact that the entire range of loss scores was reduced from 0-8 to 0-5. Also, all of the optimizer options, except for those using Adam and Adadelta, produced results that were well below a loss score value of 1.

### Final Model: 5 Hidden Layers, 8 Neurons Per Layer, Elu Activation, and NAdam Optimizer

The final model performed best with the following perameters: 1. A dense function with 8 units along with the input dimension value of 13. 2. Relu activation for the input layer, elu activation for the hidden layers, and a sigmoid activation for the output layer. 3. A total of 5 hidden layers. 4. NAdam optimization.

The results of the Adam optimizer are "generally better than every other optimization algorithms, have faster computation time, and require fewer parameters for tuning," according to www.analyticsvidhya.com. The NAdam optimizer is an improvement on the Adam optimizer that address noise within the input data. The benefits of using NAdam are demonstrrated in the test that compared optimizers. The Adam optimizer and NAdam optimizer performed well, but the Adam optimizer had more variance as it did not address noisey data as well.

The elu activation was chosen for the input and output layers after testing showed that it produced low loss scores and high accuracy. The sigmoid activation was chosen for the output layer due to the fact that it produces a value between 0 and 1. This is most appropriate for a dataset, such as this, where the output should be the probability between one value and another value.
